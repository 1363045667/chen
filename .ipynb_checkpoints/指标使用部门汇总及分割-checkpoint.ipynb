```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('D:\\思特奇\\工作文件\\最小指标库\\指标模型\\指标信息完整版v13.0（108个最小指标+147个重点衍生指标.xlsx',sheet_name='DATA使用部门')
```

```python
data
```

```python
index_id = pd.read_excel('D:\\思特奇\\工作文件\\最小指标库\\指标模型\\指标\\指标模型第五版.xlsx',usecols= [0,1])
```

```python
index_id.columns = ['指标id','指标名称']
```

```python
tmp = pd.DataFrame(columns = ['指标名称','使用部门'])
```

```python
data.iloc[12,1].split()
```

```python
for i in range(len(data)):
    tmp2 = pd.DataFrame(columns = ['指标名称','使用部门'])
    if data['使用部门'].isnull()[i] == True:
        tmp = tmp.append(data.iloc[i,:])
    else:
        tmp2['使用部门'] = data.iloc[i,1].split('、')
        tmp2['指标名称'] = data.iloc[i,0]
        tmp = tmp.append(tmp2)
```

```python
final = tmp.reset_index(drop = True).merge(index_id,on=['指标名称'],how = 'left')
```

```python
final = final[final['使用部门'].notnull()].reset_index(drop=True)
```

```python
final.to_excel('D:\\思特奇\\工作文件\\12.21 - 12.25\\12.21 最小指标库\\使用部门汇总.xlsx',index=False)
```
